```python
from stability_plots import *
import numpy as np
```

```python
G = 10  # Grid size
```

Let's start with the EZ stability coefficient plot.

```python
ez = SSYPrefs()
cp = SSYConsumption()
```

Remember original values

```python
dot_loc = ez.ψ, cp.μ_c
```

```python
psi_vec = np.linspace(1.1, 4.0, G)
mu_vec = np.linspace(0.0012, 0.0045, G)

R = np.empty((G, G))
```

```python
for i, ψ in enumerate(psi_vec):
    for j, μ_c in enumerate(mu_vec):
        ez.ψ = ψ
        cp.μ_c = μ_c
        R[i, j] = compute_ez_coef(ez, cp)
```

```python
stability_plot(R, 
               psi_vec, mu_vec, 
                "$\psi$", "$\mu_c$", 
                txt_flag='ssy',
                dot_loc=dot_loc,
                coords=(25, 25))
```

Now the spectral radius stability coefficient plot:

```python
for i, ψ in enumerate(psi_vec):
    for j, μ_c in enumerate(mu_vec):
        ez.ψ = ψ
        cp.μ_c = μ_c
        R[i, j] = compute_spec_rad_coef(ez, cp)
```

```python
stability_plot(R, 
               psi_vec, mu_vec, 
                "$\psi$", "$\mu_c$", 
               txt_flag='ssy',
               dot_loc=dot_loc,
               coords=(25, 25))
```

Now let's look at a different set of parameters:

```python
beta_vec = np.linspace(0.997, 0.9999, G)
psi_vec = np.linspace(1.25, 3.5, G)
```

```python
ez = SSYPrefs()
cp = SSYConsumption()
dot_loc = ez.β, ez.ψ
```

```python
for i, β in enumerate(beta_vec):
    for j, ψ in enumerate(psi_vec):
        ez.ψ = ψ
        ez.β = β
        R[i, j] = compute_ez_coef(ez, cp)
```

```python
stability_plot(R, 
               beta_vec, psi_vec, 
                "β", "ψ", 
                txt_flag='ssy',
                dot_loc=dot_loc)
```

And now the spec rad coefficient:

```python
for i, β in enumerate(beta_vec):
    for j, ψ in enumerate(psi_vec):
        ez.ψ = ψ
        ez.β = β
        R[i, j] = compute_spec_rad_coef(ez, cp)
```

```python
stability_plot(R, 
               beta_vec, psi_vec, 
                "β", "ψ", 
                txt_flag='ssy',
                dot_loc=dot_loc)
```
