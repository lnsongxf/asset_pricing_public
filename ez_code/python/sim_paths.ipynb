```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

from by_model import BY
```

```python
by = BY()
```

```python
z, σ = by.simulate_state(ts_length=1000)
g_c = by.simulate_consumption_given_state(z, σ)
g_d = by.simulate_dividends_given_state(z, σ)
out = g_c, g_d, z, σ
labels = "$g_c$", "$g_d$", "$z$", "$\sigma$"
```

```python
for x, la in zip(out, labels):
    fig, ax = plt.subplots()
    ax.plot(x, label=la)
    ax.legend()
    plt.show()
```
